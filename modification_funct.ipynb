```python
def modification_funct(number_of_water, total_number_of_ions, target_density, observed_density, total_atoms):
    

    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt
    

    
    f = (observed_density/target_density)**0.5
    x_old, y_old, z_old = [], [], []

    with open('conf.gro') as configuration:
        for line in configuration:
            cols = line.split()

            if len(cols)>4:
                x_old.append(float(cols[-3]))
                y_old.append(float(cols[-2]))
                z_old.append(float(cols[-1]))

    j = (4*number_of_water)  
#     total_atoms = j + total_number_of_ions
       
    x_new, y_new, z_new = [], [], []    
    ## Water
    for i in range(0, j-1,4):
        
        x_new.append(round(x_old[i] * f,3))
    
        x_new.append(round(x_old[i+1] +(f-1)*x_old[i],3))
    
        x_new.append(round(x_old[i+2] +(f-1)*x_old[i],3))
    
        x_new.append(round(x_old[i+3] +(f-1)*x_old[i],3))
 

        y_new.append(round(y_old[i] * f,3))
    
        y_new.append(round(y_old[i+1] +(f-1)*y_old[i],3))
    
        y_new.append(round(y_old[i+2] +(f-1)*y_old[i],3))
    
        y_new.append(round(y_old[i+3] +(f-1)*y_old[i],3))
    
    
        z_new.append(z_old[i])
    
        z_new.append(z_old[i+1])

        z_new.append(z_old[i+2])
    
        z_new.append(z_old[i+3])
    
    
    ## Ions
    for i in range(j,len(x_old)):
        x_new.append(round(x_old[i] * f,3))
        y_new.append(round(y_old[i] * f,3))
        z_new.append(z_old[i])
    
      
        
    data = np.array(pd.read_fwf("conf.gro", usecols=[0, 1]))

    c1 = data[1:total_atoms+1,0]

    fout = open("xxx.gro", "w")
    fout.write("NACL in Water"+"\n"+"\n")
    fout.write(str(total_atoms) + "\n")

    ind = np.linspace(1,total_atoms,total_atoms)
    IND = []
    for i in ind:
        IND.append(str(i))

        
    x, y, z = [], [], []
    
    for i,j,k in zip(x_new,y_new,z_new):
        x.append(str(i))
        y.append(str(j))
        z.append(str(k))

    for i,j,k,h,s in zip(c1,IND,x,y,z):
        fout.write(i +" "+" "+j+" "+" "+ k +" "+" "+ h +" "+" "+ s + "\n")

    with open('conf.gro') as configuration:
        for row in configuration:
            R = row.split()
            pass
        fout.write("\n"+str(round(float(R[0])*f,4)) + " "+" " + str(round(float(R[1])*f,4)) + " "+" " + str(float(R[2])))
    

    fout.close()   
    configuration.close()    
    
        
    return "xxx.gro"
    
```
